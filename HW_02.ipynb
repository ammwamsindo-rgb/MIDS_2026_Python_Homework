```python
### Q2-2 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

gasoline = int(input("Please enter a number of gallons of gasoline: "))

gallon = 3.7854
barrel_produces = 19.5
avg_price = 4.31

liter = gasoline * gallon
barrels_of_oil = gasoline / barrel_produces
price = gasoline * avg_price

print(f'{gasoline} gallons of gasoline is equivalent to {liter:.4f} liters')
print(f'Requires {barrels_of_oil:.3f} barrels of oil to produce')
print(f'The price in U.S. dollars is approximately ${price:.2f}')
      
```

```python
### Q2-3 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

a = int(input("Please provide the first number to average: "))
b = int(input("Please provide the second number to average: "))
average = int(input(
    "Please enter the type of average:\n"
    "    1. The arithmetic mean: (a + b)/2\n"
    "    2. The geometric mean: sqrt(a * b)\n"
    "    3. The root-mean-square: sqrt((a^2 + b^2)/2)\n"
    "Enter a number (1, 2, 3) from the list above: "
))

from math import sqrt

### need to define all arithmetic formulas
arithmetic_mean = (a + b) / 2
geometric_mean = sqrt(a * b)
root_mean_square = sqrt((a ** 2 + b ** 2) / 2)
              
if average == 1:
    print(f'The arithmetic mean is: {arithmetic_mean:.4f}')
elif average == 2:
    print(f'The geometric mean is: {geometric_mean:.4f}')                  
elif average == 3:
    print(f'The root-mean-square is: {root_mean_square:.4f}')
else:
    print(f'Invalid choice. Please enter 1, 2, 3.')
                     
```

```python
### Q2-4 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

income = int(input("Please enter your income: "))

first_bracket = .05
second_bracket = .1
third_bracket = .15

total_one = (first_bracket * 1000) + (second_bracket * (income - 1000))
total_two = (first_bracket * 1000) + (second_bracket * (income - 1000)) + (third_bracket * (income - 2000))

if income < 1000:
    print(f'The tax you owed is: ${first_bracket * income:.2f}')
elif income > 1000 and income < 2000:
    print(f'The tax you owed is: ${total_one:.2f}')
elif income > 2000:
    print(f'The tax you owed is: ${total_two:.2f}')
```

```python
### Q2-5 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

name = input("Enter your name: ")

entered_name = len(name) - 1

reversed_name = ""

while entered_name >= 0:
    reversed_name += name[entered_name].lower()
    entered_name -= 1
    
print()
print(reversed_name.capitalize())
if name.lower() == reversed_name.lower():
    print("Palindrome!")    
```
