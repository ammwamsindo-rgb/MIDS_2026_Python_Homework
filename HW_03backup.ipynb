```python
### Q3-2 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

numbers = input("Please enter four numbers seperated by spaces: ")

number_space = numbers.split()

a, b, c, d = map(float, number_space)

## original matrix to 4 decimal places

matrix = (
    (f"{a:.4f}", f"{b:.4f}"), 
    (f"{c:.4f}", f"{d:.4f}")
)          
print("matrix:",matrix)

det = a * d - b * c

inverse_row1 = (d/det, -b/det)
inverse_row2 = (-c/det, a/det)

## inverse matrix to 4 decimal places

inverse_matrix = (
    (f"{d/det:.4f}", f"{-b/det:.4f}"),
    (f"{-c/det:.4f}", f"{a/det:.4f}")
)

inverse = inverse_matrix
print("inverse:", inverse)
```

```python
### Q3-4 Grading Tag: Please put your entire solution in this cell. Don't edit this line.

from math import factorial

rows = int(input("Enter a row number: "))
 
# I used the Pascal's Triangle Formula to calculate = n!/((n-r)! x r!)

n = rows - 1 #this is to make sure the row number does not automatically start from 0 but 1. If the row starts from 0 row 5 will output (1 5 10 10 5 1)

for r in range(n + 1):
    ncr = factorial(n) // (factorial(r) * factorial(n - r))
    print(ncr, end=" ")
    
print()
```
